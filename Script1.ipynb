```python
import pandas as pd
import os
```

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_excel(pwd + "\\Data - Copy of Source.xlsx", sheet_name="Edited_Data")
dataset
```

```python
dataset_modified = dataset.copy()
dataset_modified
```

```python
dataset_modified.columns
```

```python
drop_list = ['StartDate', 'EndDate']
dataset_modified = dataset_modified.drop(columns=drop_list)
dataset_modified
```

```python
id_list = list(dataset_modified.columns[:6])
# id_list
```

```python
value_list = list(dataset_modified.columns[6:])
# value_list
```

```python
dataset_melted = dataset_modified.melt(id_vars=id_list, value_vars=value_list, var_name="Question+Subquestion", value_name="Answer")
dataset_melted
```

```python
questions_import = pd.read_excel(pwd + "\\Data - Copy of Source.xlsx", sheet_name="Question")
questions_import = questions_import.rename(columns={"Question + Subquestion": "Question+Subquestion"})
questions_import
```

```python
questions = questions_import.copy()
questions = questions.drop(columns=['Raw Question', 'Raw Subquestion', 'Subquestion'])
questions
```

```python
dataset_merged = pd.merge(left=dataset_melted, right=questions, how="left", left_on="Question+Subquestion", right_on="Question+Subquestion")
```

```python
dataset_merged
```

```python
respondents = dataset_merged[dataset_merged['Answer'].notna()]
respondents = respondents.groupby('Question')['ID'].nunique().reset_index()
respondents = respondents.rename(columns={"ID": "Respondents"})
respondents
```

```python
dataset_merged_two = pd.merge(left=dataset_merged, right=respondents, how="left", left_on="Question", right_on="Question")
dataset_merged_two
```

```python
same_answer = dataset_merged # [dataset_merged['Answer'].notna()]
same_answer = same_answer.groupby(['Question+Subquestion', 'Answer'])['ID'].nunique().reset_index()
same_answer = same_answer.rename(columns={"ID": "Same Answer"})
same_answer
```

```python
dataset_merged_three = pd.merge(left=dataset_merged_two, right=same_answer, how="left", left_on=['Question+Subquestion', "Answer"], right_on=['Question+Subquestion', "Answer"])
dataset_merged_three = dataset_merged_three.fillna(value = {"Same Answer": 0})
dataset_merged_three
```

```python
output = dataset_merged_three.copy()
output.to_excel(pwd + "\\Data - Final Output.xlsx", index=False)
```
